## 1. Two fair six-sided dice are rolled with the random variable $X$ representing the sum of the roll. 
### a. Give the probability mass function $f(X)$ and the cumulative-distribution function $F(X)$ as tables and plot them using R.
### b. Find $E(X)$ and $\sigma^2(X)$.


```r
a<-(2:12)
b<-c(1,2,3,4,5,6,5,4,3,2,1)
c<-b/36
d<-c 
i<-1
if (i==1){
  d[1]<-c[1]
  i<-i+1
}
while (i<12){
  d[i]<-d[i-1]+c[i]
  i<-i+1
}

print (c)
print (d)

plot (a,c)
plot (a,d)

E<-sum(a*c)
sd<-sum(((a-E)^2)*c)

print (E)
print (sd)

```
